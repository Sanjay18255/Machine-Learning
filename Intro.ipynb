```python
Question 1

To perform a matrix multiplication between a (n, k) and (k, m) matrix, we need to perform the following operations:

For each element in the (n, m) matrix, we need to perform k multiplications and k-1 additions

Since there are n rows in the first matrix and m columns in the second matrix, so here we  need to perform the above operations n*m times.

therefore,
    Total Multiplications: n*m*k
    Total Additions: n*m*(k-1)
```

```python
import numpy as np
import time

def manual_matrix_multiply(A, B):
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result

def numpy_matrix_multiply(A, B):
    return np.dot(A, B)

A = [[2, 3, 4], [5, 6, 7]]
B = [[1, 2], [3, 4], [5, 6]]

start_time = time.time()
manual_result = manual_matrix_multiply(A, B)
end_time = time.time()
print("Time taken for manual dot product:", end_time - start_time)

start_time = time.time()
numpy_result = numpy_matrix_multiply(A, B)
end_time = time.time()
print("Time taken for NumPy dot product:", end_time - start_time)

print("Manual Result:")
print(manual_result)
print("NumPy Result:")
print(numpy_result)
```

```python
import numpy as np
import time

def pass_median(A):
    n = len(A)
    for i in range(n // 2 + 1):
        max_i = i
        for j in range(i + 1, n):
            if A[j] > A[max_i]:
                max_i = j
        A[i], A[max_i] = A[max_i], A[i]
    if n % 2 == 1:
        return A[n // 2]
    else:
        return (A[n // 2 - 1] + A[n // 2]) / 2

def quickselect_median(A):
    n = len(A)
    if n % 2 == 1:
        return quickselect(A, n // 2)
    else:
        return (quickselect(A, n // 2 - 1) + quickselect(A, n // 2)) / 2

def quickselect(A, k):
    if len(A) == 1:
        return A[0]
    pivot_i = len(A)//2
    pivot = A[pivot_i]
    left = [x for i, x in enumerate(A) if x < pivot and i != pivot_i]
    middle = [x for x in A if x == pivot]
    right = [x for i, x in enumerate(A) if x > pivot and i != pivot_i]
    if k < len(left):
        return quickselect(left, k)
    elif k < len(left) + len(middle):
        return middle[0]
    else:
        return quickselect(right, k - len(left) - len(middle))

A = [12, 11, 9, 34, 0, 96]

start_time = time.time()
pass_median_result = pass_median(A.copy())
print(f"Pass median result: {pass_median_result}, Time: {time.time() - start_time:.6f} seconds")

start_time = time.time()
quickselect_median_result = quickselect_median(A.copy())
print(f"Quick select median result: {quickselect_median_result}, Time: {time.time() - start_time:.6f} seconds")

start_time = time.time()
numpy_median_result = np.median(A)
print(f"NumPy median result: {numpy_median_result}, Time: {time.time() - start_time:.6f} seconds")

print("Results match:", pass_median_result == quickselect_median_result == numpy_median_result)
```

The function f(x, y) is defined as:

f(x, y) = x^2 * y + y^3 * sin(x)

Partial derivative with respect to x: ∂f/∂x = 2 * x * y + y^3 * cos(x)

Partial derivative with respect to y: ∂f/∂y = x^2 + 3 * y^2 * sin(x)

These partial derivatives represent the gradients of the function f(x, y) for x and y.

```python
Question 6
```

```python
import sympy as sp

x, y = sp.symbols('x y')

z = (x*2)*y + (y*3) * sp.sin(x)

dz_dx = sp.diff(z, x)
dz_dy = sp.diff(z, y)

print("Partial derivative of z with respect to x:", dz_dx)
print("Partial derivative of z with respect to y:", dz_dy)
```

```python
import numpy as np
import pandas as pd
np.random.seed(0)

m = np.random.uniform(1, 2, size=(20, 5))

df = pd.DataFrame(m, columns=["a", "b", "c", "d", "e"])

sd = df.std()

max_sd = sd.max()
max_sd_col = sd[sd == max_sd].index[0]
print(f"Column with the highest standard deviation: {max_sd_col}")

mean_r = df.mean(axis=1)
min_mean_row = mean_r[mean_r == mean_r.min()].index[0]
print(f"Row with the lowest mean: {min_mean_row}")
```

```python
import numpy as np
import pandas as pd

np.random.seed(0)

m = np.random.uniform(1, 2, size=(20, 5))

df = pd.DataFrame(m, columns=["a", "b", "c", "d", "e"])

df["f"] = df['a'] + df['b'] + df['c'] + df['d'] + df['e']

df["g"] = df["f"].apply(lambda x: "LT8" if x < 8 else "GT8")

lt8 = (df["g"] == "LT8").sum()

f_lt8_sd = df[df["g"] == "LT8"]["f"].std()

f_gt8_sd = df[df["g"] == "GT8"]["f"].std()

# Print results
print(f"Number of rows where 'g' is 'LT8': {lt8}")
print(f"SD of 'f' where 'g' is 'LT8': {f_lt8_sd}")
print(f"SD of 'f' where 'g' is 'GT8': {f_gt8_sd}")
```

```python
import numpy as np

def f(l):
    a = l[0]
    ind = 0
    for i in range(1, len(l)):
        if l[i] < a:
            a = l[i]
            ind = i
    return ind

def ff(l):
    return np.argmin(l)

l = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]

print('Using normal method:', f(l), 'Using numpy:', ff(l))
```
